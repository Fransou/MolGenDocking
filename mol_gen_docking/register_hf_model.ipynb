# Register Model

```python
from mol_gen_docking.models.hf_model import  DockGenConfig, DockGenModel, DockGenProcessor, DockGenModelBase
from transformers import AutoConfig, AutoModel, AutoModelForCausalLM, AutoProcessor, AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen3-4B", trust_remote_code=True)

AutoConfig.register("dockgen", DockGenConfig)
AutoModel.register(DockGenConfig, DockGenModelBase)
AutoModelForCausalLM.register(DockGenConfig, DockGenModel)
AutoProcessor.register(DockGenConfig, DockGenProcessor)

DockGenConfig.register_for_auto_class()
DockGenModelBase.register_for_auto_class("AutoModel")
DockGenModel.register_for_auto_class("AutoModelForCausalLM")
DockGenProcessor.register_for_auto_class("AutoProcessor")

qwen = AutoModelForCausalLM.from_pretrained(
        "Qwen/Qwen3-4B"
)
dockgen = DockGenModel.from_language_model(qwen, prot_embedding_dim=320)
dockgen.push_to_hub('DockGen-Qwen3-4B', private=True, use_auth_token=True)
```

```python
tokenizer.push_to_hub('DockGen-Qwen3-4B', private=True, use_auth_token=True)
processor = DockGenProcessor(tokenizer = tokenizer)
processor.push_to_hub('DockGen-Qwen3-4B', private=True, use_auth_token=True)
```

```python
message = "Hello there! I am a test message for the DockGen model."
dockgen = AutoModelForCausalLM.from_pretrained("Franso/DockGen-Qwen3-0.6B",trust_remote_code=True,)

seq = tokenizer(message, return_tensors="pt")

sequences = seq["input_ids"]
attention_mask = seq["attention_mask"]


position_ids = attention_mask.long().cumsum(-1) - 1
position_ids.masked_fill_(attention_mask == 0, 1)

out_qwen = qwen(sequences,attention_mask=attention_mask, position_ids=position_ids)["logits"]
out_dock = dockgen(sequences,attention_mask=attention_mask, position_ids=position_ids)["logits"]
out_qwen.shape, out_dock.shape
```
