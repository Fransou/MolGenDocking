# Register Model

```python
from mol_gen_docking.models.hf_model import  DockGenConfig, DockGenModel, DockGenProcessor, DockGenModelBase
from transformers import AutoConfig, AutoModel, AutoModelForCausalLM, AutoProcessor, AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen3-0.6B", trust_remote_code=True)

AutoConfig.register("dockgen", DockGenConfig)
AutoModel.register(DockGenConfig, DockGenModelBase)
AutoModelForCausalLM.register(DockGenConfig, DockGenModel)
AutoProcessor.register(DockGenConfig, DockGenProcessor)

DockGenConfig.register_for_auto_class()
DockGenModelBase.register_for_auto_class("AutoModel")
DockGenModel.register_for_auto_class("AutoModelForCausalLM")
DockGenProcessor.register_for_auto_class("AutoProcessor")

qwen_06B = AutoModelForCausalLM.from_pretrained(
        "Qwen/Qwen3-0.6B"
)
dockgen = DockGenModel.from_language_model(qwen_06B, prot_embedding_dim=320)
dockgen.push_to_hub('DockGen-Qwen3-0.6B', private=True, use_auth_token=True)
```

```python
tokenizer.push_to_hub('DockGen-Qwen3-0.6B', private=True, use_auth_token=True)
processor = DockGenProcessor(tokenizer = tokenizer)
processor.push_to_hub('DockGen-Qwen3-0.6B', private=True, use_auth_token=True)
```
